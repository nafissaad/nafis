```python
import yfinance as yf
import pandas as pd
```

```python
def main():
    #stock Price
    tesla =yf.Ticker("TSLA")
    historical_data_tesla = tesla.history()
    print("\n Historical data \n")
    print(historical_data_tesla.head())
    
    #cashflow
    cashflow_tesla=tesla.cashflow
    print(" \n Cash flow Statement \n")
    print(cashflow_tesla)
    
    #Balancesheet
    bs_tesla = tesla.balance_sheet
    print("\n Balance sheet \n ")
    print(bs_tesla)

    return
```

```python
main()
```
