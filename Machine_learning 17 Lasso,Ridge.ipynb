```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("melb_data.csv")
df
```

```python
df.shape
```

```python
df.info(),df.nunique()
```

```python
df=df[['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount',
     'Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']]
df
```

```python
df.isna().sum()
```

```python
df = df.fillna({
        'Car': 0,
        'CouncilArea': 'Moreland',
        'BuildingArea':df.BuildingArea.mean()
    })
```

```python
df.CouncilArea.value_counts()
```

```python
df.isnull().sum()
sns.pairplot(df, hue="Price")
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X, train_y)
```

```python
reg.score(test_X, test_y)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=10000, tol=0.1)
lasso_reg.fit(train_X, train_y)
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=10000, tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

```python
ridge_reg.score(train_X, train_y)
```
