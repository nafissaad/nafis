```python
def rect_area(height = 1, length =1, isblack=True):
    area = height * length
    if isblack is True:
        print("This is a black rectangle with the area of",str(area))
    else:
        print("This is not a black rectangle with the area of",str(area))
```

```python
def process_rect(*args):
    height = args[0]
    length = args[1]
    if len(args) == 3:
        width = args[2]
        return height*length*width
    else:
        return height*length
```

```python
print(process_rect(10,20,5))
```
