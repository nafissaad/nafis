```python
balance_sheet =[]
income_statement=[]
cfs = []
years=[]
profitability_score = 0
leverag_score = 0
operating_efficiency=0
pe_ratio = 0
tickers=yf.tickers_sp500()
tickers
```

```python
def get_data(ticker):
    global balance_sheet
    global income_statement
    global cfs
    global years
    balance_sheet=yf.get_balance_sheet(ticker)
    income_statement = yf.get_income_statement(ticker)
    cfs = yf.get_cash_flow(ticker)
    years = balance_sheet.columns
```

```python
def profitability():
    global profitability_score
    #Scores #1 and 2 - net income
    net_income = income_statement[years[0]]['netIncome']
    net_income_py = income_statement[years[1]]['netIncome']
    ni_score = 1 if net_income > 0 else 0
    ni_score_2 = 1 if net_income > net_income_py else 0

    #Score #3 - operating cash flow
    op_cf = cfs[years[0]]['totalCashFromOperatingActivities']
    #print(op_cf)
    op_cf_score = 1 if op_cf > 0 else 0

    # scores 4 returnas on assets , how much it has changed over the period of time
    avg_assets = (balance_sheet[years[0]]['totalAssets']
                    + balance_sheet[years[1]]['totalAssets']) / 2
    avg_assets_py = (balance_sheet[years[1]]['totalAssets']
                    + balance_sheet[years[2]]['totalAssets']) / 2
    RoA = net_income / avg_assets
    RoA_py = net_income_py / avg_assets_py
    RoA_score = 1 if RoA > RoA_py else 0

    #Score #5 - Accruals
    total_assets = balance_sheet[years[0]]['totalAssets']
    accruals = op_cf / total_assets - RoA
    ac_score = 1 if accruals > 0 else 0

    profitability_score = ni_score + ni_score_2 + op_cf_score + RoA_score + ac_score
    print(profitability_score)
```

```python
#ni_score= negative means net loss
#ni_score_2=negative means greater loss than previous year
#op_cf_score= negative means negative cashflow
#ROA_score=negative means decreas in return assets
#ac_score=negatibe means
```

```python
for ticker in tickers[49:54]:
    try:
        get_data(ticker)
        print(ticker)
        profitability()
    except:
        print(ticker+': Something went wrong.')
```

```python
def leverage():
    global leverag_score
    #Score #6 - long-term debt ratio
    try:
        Long_Term_debt=balance_sheet[years[0]]['longTermDebt']
        total_assets = balance_sheet[years[0]]['totalAssets']
        debt_ratio = Long_Term_debt/total_assets
        debt_ratio_score=1 if debt_ratio <0.4 else 0
    except:
        debt_ratio_score = 1
    #print(debt_ratio)
    #Score #7 - Current ratio
    current_assets = balance_sheet[years[0]]['totalCurrentAssets']
    #print(current_assets)
    current_liab = balance_sheet[years[0]]['totalCurrentLiabilities']
    #print(current_liab)
    Current_ratio = current_assets/current_liab
    #print(Current_ratio)
    current_ratio_score = 1 if Current_ratio > 1.5 else 0
    #print(current_ratio_score)
        
    leverage_score = debt_ratio_score + current_ratio_score
    print(leverage_score)
    

    
```

```python
for ticker in tickers[49:54]:
    try:
        get_data(ticker)
        print(ticker)
        leverage()
    except:
        print(ticker+': Something went wrong.')
```

```python
def Operating_efficiency():
    global operating_efficiency
    #Score 9 - Gross margin
    #Score 9 Assets turnover
    gross_profit = income_statement[years[0]]['grossProfit']
    #print(gross_profit)
    gross_profit_py = income_statement[years[1]]['grossProfit']
    #print(gross_profit_py)
    revenue = income_statement[years[0]]['totalRevenue']
    #print(revenue)
    revenue_py = income_statement[years[1]]['totalRevenue']
    #print(revenue_py)
    gross_margin=gross_profit/revenue
    #print(gross_margin)
    gross_margin_py =gross_profit_py/revenue_py
    #print(gross_margin_py)    
    gross_margin_score= 1 if gross_margin >gross_margin_py else 0
    #print(gross_margin_score)
    #print(gross_margin_score)
    
    #Score 9 - Asset turnover
    avg_assets_py = (balance_sheet[years[1]]['totalAssets']
                    + balance_sheet[years[2]]['totalAssets']) / 2
    avg_assets = (balance_sheet[years[0]]['totalAssets']
                    + balance_sheet[years[1]]['totalAssets']) / 2
    asset_turnover = revenue/avg_assets
    asset_turnover_py = revenue_py/avg_assets_py
    asset_turnover_score= 1 if asset_turnover > asset_turnover_py else 0
    #print(asset_turnover_score)
    operating_efficiency=gross_margin_score+asset_turnover_score
    print(operating_efficiency)
```

```python
def pe(ticker):
    global pe_ratio
    pe_ratio = yf.get_quote_table(ticker)['PE Ratio (TTM)']
    if pe_ratio != pe_ratio: #Check if NaN
        pe_ratio = 0
    print(pe_ratio)
```

```python
for ticker in tickers[49:53]:
    try:
        get_data(ticker)
        print(ticker)
        profitability()
        leverage()
        Operating_efficiency()
        pe(ticker)
    except:
        print(ticker+': Something went wrong.')
```
