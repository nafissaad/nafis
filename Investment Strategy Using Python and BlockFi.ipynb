```python
monthly_saving= 200
rate = 0.086
#create a variable to store the time or number of time period elapsed
time = 1/12
#create a variable to store the number of times  the interest is applied per time period
num_compunded =12
# create a variable to store the number of month invested 
num_months_saving= 240# 12 month * 20 years = 240
```

```python
#create  a function to mimic the compound formula 
def compound_interest(p,r,t,n):
    #create the formula
    # A =P(1+r/n)^(tn)
    A= p*pow( (1+r/n),(t*n) )
    return A
```

```python
def investment_strategy():
    #create empty lists to store the values
    compound_list = []
    monthly_deposite_list = []
    # Add the initial deposite or the 1st month savings to the list
    compound_list.append(monthly_saving)
    monthly_deposite_list.append(monthly_saving)
    #loop through the number of months - 1
    for i in range (1, num_months_saving):# first month index is 0 , your first month is not compunded
        Amount = compound_interest(compound_list[i-1],rate,time,num_compunded)
        compound_list.append(Amount + monthly_saving)
        monthly_deposite=monthly_deposite_list[i-1]
        monthly_deposite_list.append(monthly_saving + monthly_deposite)
    return(compound_list, monthly_deposite_list)
```

```python
investment_strategy()
```

```python
monthly_deposite_list=investment_strategy()[1]
monthly_deposite_list
```

```python
compound_list=investment_strategy()[0]
total_return = compound_list[num_months_saving-1]
total_return
```

```python
#Visually Show the investments
plt.plot(compound_list,label='Comp')
plt.plot(monthly_deposite_list,label='Cash')
plt.title('Amount Gained Over Time')
plt.legend()
plt.show()
```
