```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data,columns = ["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"])
df
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
df
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df["flower_name"]= df.target.apply(lambda x: iris.target_names[x])
```

```python
iris.target_names[0]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df
```

```python
df0=df[df.flower_name=="setosa"]
df1=df[df.flower_name=="versicolor"]
df0
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(["target","flower_name"],axis="columns")
y=df.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x.values,y.values,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC(C=1,gamma=10,kernel='linear')#inside the breacket values are parameters
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.target
```

```python
dir(digits)
```

```python
df = pd.DataFrame(digits.data,digits.target)
df.head(20)
```

```python
df['target'] = digits.target
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)
```

```python
from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
```

```python
rbf_model.fit(X_train, y_train)
```

```python
rbf_model.score(X_test,y_test)
```

```python
linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)
```

```python
linear_model.score(X_test,y_test)
```
