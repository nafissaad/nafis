```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Year = []
Yearly_income = []
Yearly_Expense = []
Yearly_Investments = []
Annual_Returns= []
```

```python
#your yearly constant income
income = 35000
#yearly expense constant expense 
expense = income / 2
#Average interest is 8% per year 
interest_rate=0.08
#you invest half of your income
investment = income /2
#Calculation the annual return
annual_return = investment * interest_rate
#Getting the current year
year =2021
```

```python
#Append the first values to the lists
Year.append(year)
Yearly_income.append(income)
Yearly_Expense.append(expense)
Yearly_Investments.append(investment)
Annual_Returns.append(annual_return)
```

```python
#Loop for 'n' years
investment_year = 20
for i in range(0,investment_year-1):
    #Update the investment to be the previous investment plus the previous annual return plus half of your income
    investment = investment + annual_return + income/2
    #Update the annual return to be the current investment times the intest rate 
    annual_return = investment*interest_rate
    #Append the new data to the list
    Year.append(year+i+1)
    Yearly_income.append(income)
    Yearly_Expense.append(expense)
    Yearly_Investments.append(investment)
    Annual_Returns.append(annual_return)
```

```python
#create a DataFrame
df = pd.DataFrame()
df['Year']=Year
df['Yearly_Income'] = Yearly_income
df['Yearly_Expense'] = Yearly_Expense
df['Yearly_investmenent'] = Yearly_Investments
df['Annual_Returns'] = Annual_Returns
```

```python
df
```

```python
#Visually Show the data
plt.figure(figsize=(12,4.5))
plt.plot(df['Year'],df['Yearly_Expense'],label='Yearly_Expense')
plt.plot(df['Year'],df['Annual_Returns'],label='Annual_Returns')
plt.xticks(rotation=45)
plt.title('Retire in 10 Years')
plt.xlabel('Years')
plt.ylabel('$')
plt.xticks(df['Year'])
plt.legend()
plt.show()
```
